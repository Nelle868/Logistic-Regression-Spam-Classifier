## Evaluating Classifiers

In our models, we evaluate the accuracy of the training set, which may provide a misleading accuracy measure. In Project B1, we calculated various metrics to consider other ways of evaluating a classifier, in addition to overall accuracy. Below is a reference for these concepts.

Presumably, our classifier will be used for **filtering**, or preventing messages labeled `spam` from reaching someone's inbox. There are two kinds of errors we can make:
- **False positive (FP)**: A ham email gets flagged as spam and filtered out of the inbox.
- **False negative (FN)**: A spam email gets mislabeled as ham and ends up in the inbox.

To be clear, we label spam emails as 1 and ham emails as 0. These definitions depend both on the true labels and the predicted labels. False positives and false negatives may be of differing importance, leading us to consider more ways of evaluating a classifier in addition to overall accuracy:

**Precision**: Measures the proportion of emails flagged as spam that are actually spam. Mathematically, $\frac{\text{TP}}{\text{TP} + \text{FP}}$.

**Recall**: Measures the proportion  of spam emails that were correctly flagged as spam. Mathematically, $\frac{\text{TP}}{\text{TP} + \text{FN}}$.

**False positive rate**: Measures the proportion  of ham emails that were incorrectly flagged as spam. Mathematically, $\frac{\text{FP}}{\text{FP} + \text{TN}}$.

```python
# import libraries
# You may use any of these to create your features.
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_curve, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
import re
from collections import Counter
```

```python
# Define your processing function, processed data, and model here. 

SPAM_WORDS = [
    'free', 'click', 'money', 'offer', 'business', 'html', 'body',
    'please', 'urgent', 'guarantee', 'credit', 'order', 'limited',
    'discount', 'buy', 'save', 'winner', 'congratulations',
    'income', 'opportunity', 'investment', 'subscribe', 'remove',
    'http', 'address', 'service', 'visit', 'million', 'website',
    'available', 'shipping', 'received', 'fast', 'discover',
    'sex', 'instant', 'thousand', 'fortune', 'approved',
    'dear', 'customer', 'price', 'sale', 'cash', 'now',
    'today', 'special',
]


# ---------- Processing function ----------
def process_data(df, words=SPAM_WORDS):
    """
    Build a feature matrix from raw email data.

    Args:
        df: DataFrame with 'email' and 'subject' columns.
        words: list of words for binary presence features.

    Returns:
        2D NumPy array of shape (n_emails, n_features).
    """
    email = df['email'].fillna('')
    subject = df['subject'].fillna('')

    # Binary word-presence features (from B1's words_in_texts)
    features = pd.DataFrame(
        words_in_texts(words, email),
        columns=words
    )

    # Feature: number of HTML tags (regex from Lec 6 notes: r"<[^>]+>")
    features['num_html_tags'] = email.str.count(r'<[^>]+>')

    # Feature: number of exclamation points (suggested in project Q1 list)
    features['num_exclaim'] = email.str.count(r'!')

    # Feature: email length in characters (suggested in project Q1 list)
    features['email_len'] = email.str.len()

    # Feature: subject length in characters (suggested in project Q1 list)
    features['subj_len'] = subject.str.len()

    # Feature: is reply (suggested in project Q1 list)
    features['is_reply'] = subject.str.contains(r're:', regex=True).astype(int)

    # Feature: number of capital letters in subject (suggested in project Q1 list)
    features['num_caps_subj'] = subject.str.count(r'[A-Z]')
    
    # Feature: is forward (suggested in project Q1 list)
    features['is_forward'] = subject.str.contains(r'fw:|fwd:', regex=True).astype(int)

    # Feature: number of words in email (suggested in project Q1 list)
    features['num_words'] = email.str.split().str.len()
    
    # Feature: number of words in subject (suggested in project Q1 list)
    features['num_words_subj'] = subject.str.split().str.len()
    
    # Feature: number of digits (Lec 6 character class [0-9])
    features['num_digits'] = email.str.count(r'[0-9]')
    
    # Feature: number of dollar signs (suggested in project Q1 list)
    features['num_dollar'] = email.str.count(r'\$')
    
    # Feature: number of question marks (suggested in project Q1 list)
    features['num_question'] = email.str.count(r'\?')

    return features.to_numpy()


# ---------- Processed data ----------
X_train = process_data(train)
Y_train = np.array(train['spam'])


# ---------- Model ----------
# Cross-validation with GridSearchCV to tune the regularization
# hyperparameter, as shown in Lec 17 and described in the
# project's "Moving Forward" section.
parameters = {'C': [0.01, 0.1, 1, 10, 100]}

grid = GridSearchCV(
    estimator=LogisticRegression(max_iter=5000),
    param_grid=parameters
)
grid_result = grid.fit(X_train, Y_train)

model = grid_result.best_estimator_

print("Best C:", grid_result.best_params_['C'])
print("Mean CV scores:", grid_result.cv_results_['mean_test_score'])
print("Training accuracy:", model.score(X_train, Y_train))
```

## Question 5: ROC Curve

In most cases, we won't be able to get 0 false positives and 0 false negatives, so we have to compromise. For example, in the case of cancer screenings, false negatives are comparatively worse than false positives — a false negative means that a patient might not discover that they have cancer until it's too late. In contrast, a patient can receive another screening for a false positive.

Recall that logistic regression calculates the probability that an example belongs to a particular class. To classify an example, we say that an email is spam if our classifier gives it $\ge 0.5$ probability of being spam. However, **we can adjust that cutoff threshold**. How does the ambiguity in our labeled data (spam or ham) affect our understanding of the model's predictions and the way we measure/evaluate our model's performance?

The ambiguity in our labeled data means our evaluation metrics like accuracy, precision, and recall are only as reliable as the ground truth labels themselves. If some emails were labeled inconsistently or based on subjective judgment, then a prediction the model gets "wrong" might actually be a reasonable classification that disagrees with a questionable label. This can inflate our false positive and false negative counts in misleading ways. This also means a model with lower reported accuracy could actually be making sensible predictions, and we should treat performance metrics as estimates of model quality rather than exact measurements of how well the model captures real spam versus ham.
